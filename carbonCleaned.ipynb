```python
import pandas as pd

ciaLink1='https://www.cia.gov/the-world-factbook/field/carbon-dioxide-emissions/country-comparison'
carbon=pd.read_html(ciaLink1, # link
                        header=0, # where is the header?
                        flavor='bs4')[0] # and which table?
```

```python
carbon
```

```python
#1.
byeColumns=['Rank']
carbon.drop(columns=byeColumns,inplace=True)
carbon
```

```python
#2.
carbon.rename(columns={"Country": "País", "metric tonnes of CO2": "Toneladas_CO2", "Date of Information": "Fecha"   }, inplace=True )
carbon
```

```python
#3.
carbon.País.str.contains(pat=r'\^',regex=True)
```

```python
carbon.Fecha.str.contains(pat=r'\^',regex=True)
```

```python
#4.
carbon.Fecha.str.contains(pat=r'\^',
                          regex=True)
```

```python
carbon2=carbon.copy()

byeColumns=['País', 'Fecha']
carbon2.drop(columns=byeColumns, inplace=True)
carbon2
```

```python
carbon2['Toneladas_CO2'] = carbon2['Toneladas_CO2'].astype('object')
carbon.Fecha.str.contains(pat=r'\^',
                          regex=True)
```

```python
#5.
carbon.columns = carbon.columns.str.replace(' ', '_', regex=False)
print("\nPaís:", "\nToneladas_CO2:", "\nFecha:")
print(carbon.columns)
```
