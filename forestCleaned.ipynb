```python
import pandas as pd

ciaLink2='https://www.cia.gov/the-world-factbook/field/revenue-from-forest-resources/country-comparison/'
carbon_=pd.read_html(ciaLink2, # link
                        header=0, # where is the header?
                        flavor='bs4')[0] # and which table?
```

```python
carbon_
```

```python
#1.
byeColumns=['Rank']
carbon_.drop(columns=byeColumns,inplace=True)
carbon_
```

```python
#2.
carbon_.rename(columns={"Country": "País", "% of GDP": "PercentageGDP", "Date of Information": "Fecha"   }, inplace=True )
carbon_
```

```python
#3.
carbon_.País.str.contains(pat=r'\^',regex=True)
```

```python
carbon_.Fecha.str.contains(pat=r'\^',regex=True)
```

```python
#4.
carbon_.Fecha.str.contains(pat=r'\^',
                          regex=True)
```

```python
carbon_2=carbon_.copy()

byeColumns=['País', 'Fecha']
carbon_2.drop(columns=byeColumns, inplace=True)
carbon_2
```

```python
carbon_2['PercentageGDP'] = carbon_2['PercentageGDP'].astype('object')
carbon_.Fecha.str.contains(pat=r'\^',
                          regex=True)
```
